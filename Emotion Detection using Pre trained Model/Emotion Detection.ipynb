```python
import cv2
import matplotlib.pyplot as plt
from deepface import DeepFace
# https://github.com/serengil/deepface
```

```python
img = cv2.imread('sad.jpg')
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.show()
```

```python
predictions = DeepFace.analyze(img)
predictions
```

```python
text = predictions['dominant_emotion']
x,y,w,h = predictions['region'].values()


cv2.rectangle(img,(x,y), (x+w,y+h),(255,0,0),3) 
cv2.putText(img, text, (round((x-10+w)/2),y-10),fontFace=2,fontScale=2, color=(255,0,0), thickness=2)
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```

```python
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    predictions = DeepFace.analyze(frame)
    text = predictions['dominant_emotion']
    x,y,w,h = predictions['region'].values()


    cv2.rectangle(frame,(x,y), (x+w,y+h),(255,0,0),3) 
    cv2.putText(frame, text, (round((x-10+w)/2),y-10),fontFace=2,fontScale=2, color=(255,0,0), thickness=2)
    cv2.imshow('My Frame', frame)
    
    if cv2.waitKey(1) & 0xFF==ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
